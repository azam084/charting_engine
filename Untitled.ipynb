```python
import pygal
import pandas as pd
from pygal.style import Style
import math
import cairosvg

 

from pygal.util import alter, decorate, ident, swap

 

data = pd.read_excel(r'C:\Users\yamna\Desktop\CarPrices\Pygal\CostOfLivingIndexYear.xlsx')

 

dff=data[:]
df=pd.DataFrame(dff)

 


def round_up(n, decimals=0):
    multiplier = 10 ** decimals
    return math.ceil(n * multiplier) / multiplier
NumberOfBars = df.shape[0]
isQtr_on_x=1 # if isQtr_on_x is 1 then

 

def calculateNumberOfMajors():
    if isQtr_on_x ==1:
        numberOfMajors=int(NumberOfBars) if (NumberOfBars<14) else int(round_up(NumberOfBars/2)) if (NumberOfBars>=14 and NumberOfBars<= 25) else 13  #display number of qtrs 
        return numberOfMajors
    else:
        numberOfMajors=int(NumberOfBars) if (NumberOfBars<21) else int(round_up(NumberOfBars/2)) if (NumberOfBars>=21 and NumberOfBars<= 29) else 5  #display number of years 
        return numberOfMajors

 

 

# numberOfMajors=int(NumberOfBars) if (NumberOfBars<21) else int(round_up(NumberOfBars/2)) if (NumberOfBars>=21 and NumberOfBars<= 29) else 5
# numberOfMajors=int(NumberOfBars) if (NumberOfBars<21)  else 5

 


x_lbl_rotation=0
lang='english'
# lang='arabic'
# bar_graph_title='هنا سيكون لدينا عنوان الرسم البياني'
# bar_graph_title='هذا عنوان جديد وأنا على وشك وضع عنوان جديد هنا'
bar_graph_title='title here'

 

class MyBarChart(pygal.Bar):

 

    def _init_(self, *args, **kwargs):
        super(MyBarChart, self)._init_(*args, **kwargs)

 

    def _make_title(self):
        """Make the title"""
        if self._title:
            for i, title_line in enumerate(self._title, 1):
                self.svg.node(
                    self.nodes['title'],
                    'text',
                    class_='title plot_title', 
                    x=self.margin_box.left if lang == 'english'else self.width - self.margin_box.right,   
#                     x=self.width - self.margin_box.right, # Modified by default The x-coordinate of each line is set to the center of the plot (half the width of the plot), 
                    y=i * (self.style.title_font_size + self.spacing)  # by default the y-coordinate is calculated based on the line number, font size, and spacing.
                ).text = title_line


    def _tooltip_and_print_values(
            self, serie_node, serie, parent, i, val, metadata, x, y, width,
            height
    ):
        transpose = swap if self.horizontal else ident
        # x_center, y_center = transpose((x + width / 2, y + height / 2)) before
        x_center, y_center = transpose((x + width/2 , (y- height/4))) # here i have changed to make tooltip at the top of bar
        x_top, y_top = transpose((x + width, y + height))
        x_bottom, y_bottom = transpose((x, y))
        if self._dual:
            v = serie.values[i][0]
        else:
            v = serie.values[i]
        sign = -1 if v < self.zero else 1
        self._tooltip_data(
            parent, val, x_center, y_center, "centered", self._get_x_label(i)
        )

 

        if self.print_values_position == 'top':
            if self.horizontal:
                x = x_bottom + sign * self.style.value_font_size / 2
                y = y_center
            else:
                x = x_center
                y = y_bottom - sign * self.style.value_font_size / 2
        elif self.print_values_position == 'bottom':
            if self.horizontal:
                x = x_top + sign * self.style.value_font_size / 2
                y = y_center
            else:
                x = x_center
                y = y_top - sign * self.style.value_font_size / 2
        else:
            x = x_center
            y = y_center
        self._static_value(serie_node, val, x, y, metadata, "middle")

 



    # Define custom styles for the chart
custom_style_basicBarGraph = Style(
        background='#F1F1F1',
        plot_background='transparent',
        foreground='#777575',
        foreground_strong='black',
        font_family='googlefont:Lato',
        opacity='.7',
        opacity_hover='1',
        transition='400ms ease-in',
        value_label_font_family='',
        tooltip_font_family='',
        title_font_family='',
        label_font_size=10, # size for the x and y axis ticks or u may say labels/values
        colors=['#F1953C', '#E8537A'], #orange color , yellow color  for bars 
        value_colors=['black',] ,# color for values displayig at the top of bar

 

    )
basicBarGraphConfig = pygal.Config()

 

 

    # make guide lines lighter, clashes with printed values.
basicBarGraphConfig.css.append('''inline:
      g.guides path.guide.line{
        stroke: rgb(168,168,168) !important;
      }''')

 

    # If we hover over the label or the line, make line full black.and bold the text
basicBarGraphConfig.css.append('''inline:
      g.guides:hover path.guide.line {
        font-weight: bold;
      }''')

 


basicBarGraphConfig.css.append('''inline:
      g.tooltip rect {
        fill: rgb(0,0,0) !important;
      }''')

 

basicBarGraphConfig.css.append('''inline:
      g.tooltip  .value{
      font-size: 10px  !important;
       fill: rgb(255,255,255) !important;
      font-weight: bold !important;
      font-family: OpenSansRegular !important;

 

      }''')
# font-family: monospace
basicBarGraphConfig.css.append('''inline:
      g.tooltip  text{
      font-size: 10px  !important;
       fill: rgb(255,255,255) !important;
        font-family: OpenSansRegular;
      }''')

 

basicBarGraphConfig.css.append('''inline:
      g.tooltip .x_label {
       /* display:none;*/
      }''')

 

 


def drawBarGraphForMe(chartTitle,x_data,y_data,graphSaveAs='NewBargraph',language='english',isQtr_on_xaxis=0,isRankingChart=0):
    global lang,x_lbl_rotation,isQtr_on_x
    lang=language
    isQtr_on_x=isQtr_on_xaxis
    x_lbl_rotation=0 if isRankingChart ==0 else -40
    if lang == 'english':
        basicBarGraphConfig.css.append('''inline:
      g.titles text.title.plot_title {
        /*font-size: 16px !important;*/
        font-weight: bold;
        text-anchor: start !important;  
      }''')


    else:
        basicBarGraphConfig.css.append('''inline:
        g.titles text.title.plot_title {
            /*font-size: 16px !important;*/
            font-weight: bold;
            text-anchor: end !important;  
        }''')
    bar_graph=MyBarChart(
        
        basicBarGraphConfig,
        no_data_text='No Data Avaiable',
        title=chartTitle,
        width=600,
        height=250,
        style=custom_style_basicBarGraph,
        show_legend =False,
        show_minor_y_labels=True,  
        show_major_y_labels=False,  
        show_major_x_labels=True,
        show_minor_x_labels=False,
        x_label_rotation=x_lbl_rotation,  #-40  for company name rotation zero default 
        x_labels_major_count=calculateNumberOfMajors(),  #change here numberOfMajors    
        truncate_label= 30,
        show_y_guides=True,
        show_x_guides=False,
        # print_values=True,
        # print_values_position='top',
    )

 

    # bar_graph._x_labels_major=df['FiscalPeriod']
    bar_graph.add('',y_data)
    bar_graph.x_labels=x_data
    bar_graph.render_to_file("{graphName}.svg".format(graphName=graphSaveAs))
    bar_graph.render_to_png("{graphName}.png".format(graphName=graphSaveAs))

 

    
# drawBarGraphForMe(bar_graph_title,df['FiscalPeriod'],df['3MonthSAIBOR'],'BargraphNew2.svg')
drawBarGraphForMe(bar_graph_title,df['FiscalPeriod'],df['CostOfLivingIndex'],language='english',graphSaveAs='aBarGraphFinal',isRankingChart=0,isQtr_on_xaxis=0)
```
